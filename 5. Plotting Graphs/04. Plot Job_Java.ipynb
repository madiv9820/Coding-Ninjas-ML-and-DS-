```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

dataset = pd.read_csv('amazon_jobs_dataset.csv', skipinitialspace = True, encoding = 'utf-8')
dataset['Year'] = [date[-4:] for date in dataset.Posting_date]
dataset['Contains_Java'] = [True if 'java' in str(qualification).lower() else False for qualification in dataset['BASIC QUALIFICATIONS']]

years = dataset.Year[dataset['Contains_Java'] == True]

no_of_Java_Jobs_Year_Wise = np.array([(year, no_of_jobs) for year, no_of_jobs in Counter(years).items()], dtype = int)
np.sort(no_of_Java_Jobs_Year_Wise)
no_of_Java_Jobs_Year_Wise = no_of_Java_Jobs_Year_Wise[::-1]

years = no_of_Java_Jobs_Year_Wise[:, 0]
no_of_Jobs = no_of_Java_Jobs_Year_Wise[:, 1]

plt.scatter(years, no_of_Jobs)
plt.plot(years, no_of_Jobs)
plt.ylabel('No of Jobs', fontsize = 10)
plt.xlabel('Year', fontsize = 10)
plt.title('Java Jobs V/S Year', fontsize = 15)
plt.plot()

for index in range(len(years)): print(years[index], no_of_Jobs[index])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('amazon_jobs_dataset.csv',skipinitialspace = True,encoding='UTF-8')
date = dataset.Posting_date
qual = dataset['BASIC QUALIFICATIONS']
years = []
for i in range(len(date)):
    if 'Java' in str(qual[i]) or 'java' in str(qual[i]):
        years.append(int(date[i][-4:]))
years = np.array(years)
year = np.sort(np.unique(years))
freq = []
for y in year:
    freq.append(len(np.where(years == y)[0]))
plt.scatter(year,freq, color = ['red','green','yellow','blue','pink','magenta','purple'])
plt.plot(year,freq, color = 'grey',linewidth = 0.5)
plt.title('Trend of Job (2012-2018)')
plt.xlabel('Year')
plt.ylabel('Jobs')
plt.plot()
for i in range(len(year)):
    print(year[i],freq[i])
```
