```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

dataset = pd.read_csv('amazon_jobs_dataset.csv', skipinitialspace = True, encoding = 'utf-8')
months = [date.split()[0] for date in dataset.Posting_date]

no_of_Jobs_Month_Wise = np.array([(month, no_of_Jobs) for month, no_of_Jobs in Counter(months).items()])
months = no_of_Jobs_Month_Wise[:, 0]
no_of_Jobs = np.array(no_of_Jobs_Month_Wise[:, 1], dtype = int)

plt.bar(months, no_of_Jobs, edgecolor = 'black')
plt.xticks(rotation = 40)
plt.yticks(np.arange(0, 1101, 100))
plt.xlabel('Months', fontsize = 10)
plt.ylabel('No of Jobs', fontsize = 10)
plt.title('Jobs vs Months')
plt.show()

for index in range(len(months)):
    print(months[index], no_of_Jobs[index])
```
