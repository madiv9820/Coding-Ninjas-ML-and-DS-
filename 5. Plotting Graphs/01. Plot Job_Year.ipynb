```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

dataset = pd.read_csv('amazon_jobs_dataset.csv', skipinitialspace = True, encoding = 'utf-8')
years = [int((date[-1:-5:-1])[::-1]) for date in dataset.Posting_date]

no_of_Jobs_Year_Wise = np.array([(year, jobs) for year, jobs in Counter(years).items()])
np.sort(no_of_Jobs_Year_Wise, axis = 0)

years = no_of_Jobs_Year_Wise[:, 0]
no_of_jobs = no_of_Jobs_Year_Wise[:, 1]

plt.scatter(years, no_of_jobs)
plt.plot(years, no_of_jobs)
plt.yticks(np.arange(0, 1801, 100))
plt.xlabel('Year')
plt.ylabel('No of Jobs')
plt.title('Jobs vs Year', fontsize = 15)
plt.grid()
```
