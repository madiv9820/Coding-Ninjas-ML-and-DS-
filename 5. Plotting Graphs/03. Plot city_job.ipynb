```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

dataset = pd.read_csv('amazon_jobs_dataset.csv', skipinitialspace = True, encoding = 'utf-8')
dataset.dropna(inplace = True)
dataset['Country'] = [location.split(', ')[0] for location in dataset['location']]
dataset['City'] = [location.split(', ')[-1] for location in dataset.location]

cities = dataset.City[dataset.Country == 'IN']
no_of_Jobs_City_Wise = np.array([(city, no_of_Jobs) for city, no_of_Jobs in Counter(cities).items()])

cities = no_of_Jobs_City_Wise[:, 0]
no_of_Jobs = np.array(no_of_Jobs_City_Wise[:, 1], dtype = int)
total_Jobs = np.sum(no_of_Jobs)

plt.pie(no_of_Jobs, autopct = '%.2f%%', labels = cities, startangle = 60)
plt.show()

for index in range(len(cities)):
    print(cities[index], '{:.2f}'.format(no_of_Jobs[index]/total_Jobs * 100))
```
